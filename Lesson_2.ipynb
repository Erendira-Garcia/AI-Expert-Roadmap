##Lematización

La lematización es el proceso que transforma una palabra a su forma base que se denomina lema o raíz. Se basa en reglas gramaticales y en el conocimiento de la estructura y la morfología de las palabras en un idioma determinado.

image.png

```python
lemas=[]
lemmatizer = WordNetLemmatizer()
for token in tokens_en:
  lemas.append(lemmatizer.lemmatize(token.lower()))
text_in_lemmas = ' '.join(lemas)
print('Texto en lemas: ', text_in_lemmas)
```

```python
lemmatizer.lemmatize("can't")
```

```python
lemas2=[]
for token in tokens_en2:
  lemas2.append(lemmatizer.lemmatize(token.lower()))
text_in_lemmas2 = ' '.join(lemas2)
print('Texto en lemas: ', text_in_lemmas2)
```

*   ADJ: adjective
*   ADP: adposition
*   ADV: adverb
*   AUX: auxiliary
*   CCONJ: coordinating conjunction
*   DET: determiner
*   INTJ: interjection
*   NOUN: noun
*   NUM: numeral
*   PART: particle
*   PRON: pronoun
*   PROPN: proper noun
*   PUNCT: punctuation
*   SCONJ: subordinating conjunction
*   SYM: symbol
*   VERB: verb
*   X: other

##Eliminar *Stopwords*

Las palabras sin significado o *Stopwords* son palabras que suelen eliminarse en las aplicaciones del PLN debido a que reducen el ruido que pueden generar, además de que aumentan la dimensión del vocabulario y esto genera un mayor consumo de recursos computacionales.

Sin embargo, es importante saber que no en todos los casos se eliminan estas palabras, como en algunas aplicaciones del análisis de sentimientos o cuando se requiere conocer la intención o el contexto de un texto.

image.png

####Eliminando *stopwords* con NLTK

```python
stop_words = set(stopwords.words("english"))
stopwords_es = set(stopwords.words('spanish'))
```

```python
palabras_filtradas = [palabra for palabra in tokens_es if palabra.lower() not in stopwords_es]
texto_filtrado = ' '.join(palabras_filtradas)
print(texto_filtrado)
```

```python
filtered_sentence = [word for word in tokens_en if word.casefold() not in stop_words]
filtered_sentence = " ".join(filtered_sentence)
print(filtered_sentence)
```

```python
filtered_sentence = [word for word in tokens_en2 if word.casefold() not in stop_words]
filtered_sentence = " ".join(filtered_sentence)
print(filtered_sentence)
```

####Eliminando *stopwords* con Spacy

```python
tokens_sin_stopwords_es = [token.text for token in spacy_doc_es if not token.is_stop]
print(' '.join(tokens_sin_stopwords_es))
```

```python
tokens_sin_stopwords_en = [token.text for token in spacy_doc_en if not token.is_stop]
print(' '.join(tokens_sin_stopwords_en))
```

##Modelos de Representación

En el PLN los modelos de representación son enfoques que se utilizan para representar el significado y la estructura del lenguaje humano cuyo objetivo principal es permitir que las computadoras comprendan y procesen el lenguaje humano de una forma efectiva.

En esta ocasión utilizaremos 2 de los modelos que existen y son:

*   Bolsa de palabras (BoW)
*   *Word2Vec*

####Bolsa de Palabras (BoW)

El modelo de bolsa de palabras es uno de los más básicos en el PLN, ya que representa un texto como un conjunto de palabras sin considerar su orden ni su relación. Cada palabra se trata de forma independiente y se crea un vector que indica la presencia o ausencia de cada palabra en el texto.

image.png

####Utilizando Sklearn para generar los vectores de la bolsa de palabras

La librería de *Scikit-learn* es una librería enfocada en *Machine Learning*. No solo tiene métodos enfocados al PLN, ya que tambien contiene diversos algoritmos para aplicar aprendizaje automático (*Machine Learning*).

*   Scikit Learn 🚀 https://scikit-learn.org/stable/user_guide.html

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documentos = [
    'Este es el primer documento',
    'Ahora este es el segundo documento',
    'Este documento es el tercer documento',
    '¿Este documento es el primer primer primer primer documento?',
]
```

```python
vectorizer = CountVectorizer()
```

```python
X = vectorizer.fit_transform(documentos)
```

```python
vocabulario = vectorizer.get_feature_names_out()
print(vocabulario)
```

```python
print("\nBag of Words:")
for i, doc in enumerate(documentos):
    print(f"Documento {i+1}:")
    print(X.toarray()[i])
    print()
```

####Word2Vec

Word2Vec es un modelo de representación vecorial cuyo objetivo principal es capturar las relaciones semánticas y las similitudes entre palabras en un corpus de texto.

Word2Vec se basa en la idea de que el significado de una palabra está relacionado con el contexto en el que aparece. Entonces, en lugar de representar las palabras como unidades discretas (como BoW), Word2Vec utiliza una representación vectorial distribuida en la que las palabras se representan como vectores de números reales.

Word2Vec utiliza dos arquitecturas principales: Continuous Bag of Words (CBOW) y Skip-gram.

En la arquitectura **CBOW**, el objetivo es predecir una palabra objetivo a partir de un contexto de palabras circundantes. Por otro lado, la arquitectura **Skip-gram** tiene como objetivo predecir el contexto de palabras circundantes a partir de una palabra objetivo.

####Entrenando Word2Vec

```python
oraciones = [
    "Mi gato come pescado",
    "Mi perro duerme en su cama",
    "Mi hermano juega con su amigo",
    'El gato es un felino'
]
```

```python
oraciones_tokenizadas = [nltk.word_tokenize(oracion.lower()) for oracion in oraciones]
```

```python
modelo = Word2Vec(oraciones_tokenizadas, vector_size=100, window=5, min_count=1, workers=4, sg=0)
modelo.train(oraciones_tokenizadas, total_examples=len(oraciones_tokenizadas), epochs=100)
```

```python
gato_vec = modelo.wv['gato']
print("Vector de la palabra 'gato':\n", gato_vec)
```

```python
#Word2Vec-English
sents = [
    "My cat eats fish",
    "My dog sleeps in his bed",
    "My brother plays with his friend",
    "The cat is a feline"
]
```

```python
sents_tokenized = [nltk.word_tokenize(sent.lower()) for sent in sents]
```

```python
model = Word2Vec(sents_tokenized, vector_size=100, window=5, min_count=1, workers=4, sg=0)
```

```python
cat_vec = model.wv['cat']
print("Vector de la palabra 'cat':\n", cat_vec)
```
