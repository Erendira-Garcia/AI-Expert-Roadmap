##Eliminar *Stopwords*

Las palabras sin significado o *Stopwords* son palabras que suelen eliminarse en las aplicaciones del PLN debido a que reducen el ruido que pueden generar, adem谩s de que aumentan la dimensi贸n del vocabulario y esto genera un mayor consumo de recursos computacionales.

Sin embargo, es importante saber que no en todos los casos se eliminan estas palabras, como en algunas aplicaciones del an谩lisis de sentimientos o cuando se requiere conocer la intenci贸n o el contexto de un texto.

image.png

####Eliminando *stopwords* con NLTK

```python
stop_words = set(stopwords.words("english"))
stopwords_es = set(stopwords.words('spanish'))
```

```python
palabras_filtradas = [palabra for palabra in tokens_es if palabra.lower() not in stopwords_es]
texto_filtrado = ' '.join(palabras_filtradas)
print(texto_filtrado)
```

####Bolsa de Palabras (BoW)

El modelo de bolsa de palabras es uno de los m谩s b谩sicos en el PLN, ya que representa un texto como un conjunto de palabras sin considerar su orden ni su relaci贸n. Cada palabra se trata de forma independiente y se crea un vector que indica la presencia o ausencia de cada palabra en el texto.

####Word2Vec

Word2Vec es un modelo de representaci贸n vecorial cuyo objetivo principal es capturar las relaciones sem谩nticas y las similitudes entre palabras en un corpus de texto.

Word2Vec se basa en la idea de que el significado de una palabra est谩 relacionado con el contexto en el que aparece. Entonces, en lugar de representar las palabras como unidades discretas (como BoW), Word2Vec utiliza una representaci贸n vectorial distribuida en la que las palabras se representan como vectores de n煤meros reales.

Word2Vec utiliza dos arquitecturas principales: Continuous Bag of Words (CBOW) y Skip-gram.

En la arquitectura **CBOW**, el objetivo es predecir una palabra objetivo a partir de un contexto de palabras circundantes. Por otro lado, la arquitectura **Skip-gram** tiene como objetivo predecir el contexto de palabras circundantes a partir de una palabra objetivo.

####Entrenando Word2Vec

```python
oraciones = [
    "Mi gato come pescado",
    "Mi perro duerme en su cama",
    "Mi hermano juega con su amigo",
    'El gato es un felino'
]
```

```python
oraciones_tokenizadas = [nltk.word_tokenize(oracion.lower()) for oracion in oraciones]
```

```python
modelo = Word2Vec(oraciones_tokenizadas, vector_size=100, window=5, min_count=1, workers=4, sg=0)
modelo.train(oraciones_tokenizadas, total_examples=len(oraciones_tokenizadas), epochs=100)
```

```python
gato_vec = modelo.wv['gato']
print("Vector de la palabra 'gato':\n", gato_vec)
```

```python
#Word2Vec-English
sents = [
    "My cat eats fish",
    "My dog sleeps in his bed",
    "My brother plays with his friend",
    "The cat is a feline"
]
```

```python
sents_tokenized = [nltk.word_tokenize(sent.lower()) for sent in sents]
```

```python
model = Word2Vec(sents_tokenized, vector_size=100, window=5, min_count=1, workers=4, sg=0)
```

```python
cat_vec = model.wv['cat']
print("Vector de la palabra 'cat':\n", cat_vec)
```
