```python
file_path = '/content/drive/My Drive/amazon_prime_users.csv'
df = pd.read_csv(file_path)

# Mostrar las primeras filas del dataframe para verificar
df.head()
```

```python
# Conteo de valores faltantes por columna
print("\n VALORES FALTANTES POR COLUMNA \n")
print(df.isnull().sum())

# Conteo de valores faltantes por fila
print("\n VALORES FALTANTES POR FILA \n")
print(df.isnull().sum(axis=1))
```

```python
# Resumen estad√≠stico de las variables num√©ricas
print("\n RESUMEN ESTADISTICO \n")
df.describe()
```

```python
# Convertir 'Date of Birth' a formato de fecha
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'], errors='coerce')

# Convertir 'Membership Start Date' y 'Membership End Date' a formato de fecha
df['Membership Start Date'] = pd.to_datetime(df['Membership Start Date'], errors='coerce')
df['Membership End Date'] = pd.to_datetime(df['Membership End Date'], errors='coerce')

# Verificar fechas err√≥neas
invalid_birth_dates = df[df['Date of Birth'].isnull()]
invalid_start_dates = df[df['Membership Start Date'].isnull()]
invalid_end_dates = df[df['Membership End Date'].isnull()]

print("Fechas de nacimiento inv√°lidas:")
print(invalid_birth_dates)
print("Fechas de inicio de membres√≠a inv√°lidas:")
print(invalid_start_dates)
print("Fechas de fin de membres√≠a inv√°lidas:")
print(invalid_end_dates)
```

```python
print("Fechas de nacimiento inv√°lidas:")
print(invalid_birth_dates)
```

```python
# Crear una columna para la duraci√≥n de la membres√≠a en d√≠as
df['Membership Duration'] = (df['Membership End Date'] - df['Membership Start Date']).dt.days

# Verificar la nueva columna
print(df[['Membership Start Date', 'Membership End Date', 'Membership Duration']].head())

# Convertir 'Feedback/Ratings' a una escala del 1 al 5, si es necesario
df['Feedback/Ratings'] = df['Feedback/Ratings'].clip(upper=5)

# Mostrar un resumen de estad√≠sticas descriptivas para verificar
print(df.describe())
```

```python
# Visualizaci√≥n de la distribuci√≥n de la columna 'Feedback/Ratings'
plt.figure(figsize=(10, 6))
sns.histplot(df['Feedback/Ratings'], kde=True, bins=10)
plt.title('Distribuci√≥n de Ratings')
plt.show()
```

```python
# Visualizaci√≥n de la distribuci√≥n de la columna 'Membership Duration'
plt.figure(figsize=(10, 6))
sns.histplot(df['Membership Duration'], kde=True, bins=30)
plt.title('Distribuci√≥n de Duraci√≥n de Membres√≠a')
plt.show()
```

```python
# Boxplot de 'Membership Duration' por 'Subscription Plan'
plt.figure(figsize=(14, 8))
sns.boxplot(x='Subscription Plan', y='Membership Duration', data=df)
plt.title('Duraci√≥n de Membres√≠a por Plan de Suscripci√≥n')
plt.xticks(rotation=45)
plt.show()
```

```python
# Countplot de 'Gender'
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Distribuci√≥n de G√©nero')
plt.show()
```

```python
# Crear una columna 'Year' a partir de 'Membership Start Date'
df['Year'] = pd.DatetimeIndex(df['Membership Start Date']).year

# Series temporal de nuevos miembros por a√±o
new_members_per_year = df['Year'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
new_members_per_year.plot(kind='line')
plt.title('Nuevos Miembros por A√±o')
plt.xlabel('A√±o')
plt.ylabel('N√∫mero de Nuevos Miembros')
plt.grid(True)
plt.show()
```

```python
# Segmentaci√≥n de usuarios por Plan de Suscripci√≥n y G√©nero
subscription_gender_group = df.groupby(['Subscription Plan', 'Gender']).size().unstack()

plt.figure(figsize=(12, 8))
subscription_gender_group.plot(kind='bar', stacked=True)
plt.title('Distribuci√≥n de G√©nero por Plan de Suscripci√≥n')
plt.xlabel('Plan de Suscripci√≥n')
plt.ylabel('N√∫mero de Usuarios')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
from datetime import datetime

# Definir una funci√≥n para calcular la edad
def calculate_age(birthdate):
    today = datetime.today()
    age = today.year - birthdate.year - ((today.month, today.day) < (birthdate.month, birthdate.day))
    return age

# Aplicar la funci√≥n a la columna 'Date of Birth' y crear la nueva columna 'Age'
df['Age'] = df['Date of Birth'].apply(lambda x: calculate_age(x) if pd.notnull(x) else None)

# Verificar las primeras filas de la nueva columna
print(df[['Date of Birth', 'Age']].head())
```

```python
# Aseg√∫rate de que la columna 'Usage Frequency' est√° en un formato adecuado para el an√°lisis
# Por ejemplo, si es una columna categ√≥rica, puedes necesitar convertirla a num√©rica

# Supongamos que 'Usage Frequency' es una variable num√©rica
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Usage Frequency', data=df)
plt.title('Frecuencia de Uso por Edad')
plt.xlabel('Edad')
plt.ylabel('Frecuencia de Uso')
plt.grid(True)
plt.show()
```

```python
# Verificar los valores √∫nicos en la columna 'Usage Frequency'
print(df['Usage Frequency'].unique())
```

```python
from sklearn.preprocessing import LabelEncoder

# Crear un LabelEncoder
label_encoder = LabelEncoder()

# Ajustar el LabelEncoder a la columna 'Usage Frequency' y transformarla
df['Usage Frequency Numeric'] = label_encoder.fit_transform(df['Usage Frequency'])

# Verificar las primeras filas para asegurarse de que la columna se ha transformado correctamente
print(df[['Usage Frequency', 'Usage Frequency Numeric']].head())
```

```python
# Seleccionar solo las columnas num√©ricas del DataFrame
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Mostrar las primeras filas del DataFrame num√©rico para verificar
print(numeric_df.head())
```

```python
# Calcular la matriz de correlaci√≥n solo para las columnas num√©ricas
correlation_matrix = numeric_df.corr()

# Mostrar la matriz de correlaci√≥n
print(correlation_matrix)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un heatmap para visualizar la matriz de correlaci√≥n
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlaci√≥n para Columnas Num√©ricas')
plt.show()
```
